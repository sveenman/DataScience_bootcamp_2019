## Sampling data
Let's create some datasets bij sampling from some distributions. We use the package numpy to sample data. More information on numpy can be found here: http://docs.scipy.org/doc/numpy/reference/routines.random.html

```python
# we need to tell the notebook server to plot inline
%matplotlib inline

import numpy as np
import pandas as pd
#draw 1000 values from a normal distribution with mean 40 and std dev 20.
vData = np.random.normal(size=10000, loc=40, scale=20)
import matplotlib.pyplot as plt
plt.hist(vData,bins=100)
plt.show()
```

**What happens when you decrease the sample size?**

**Also try changing the distribution to another distribution found in the link above (try Laplace for instance)**

```python
import numpy as np
import pandas as pd
#draw 10000 values from a normal distribution with mean 40 and std dev 20.
vData = np.random.laplace(size=10000, loc=40, scale=20)
import matplotlib.pyplot as plt
plt.hist(vData,bins=100)
plt.show()
```

##Plotting
We will look at two different packages for plotting in iPython. Matplotlib (http://matplotlib.org/api/pyplot_api.html) and pandas dataframe (http://pandas.pydata.org/pandas-docs/stable/api.html).

There are different types of plots all giving different information. 

Today we will look at three commonly used plots:

1. Scatter plot
2. Histogram
3. Boxplot

### 1. Scatter plot

A scatterplot shows the relation between two variables x and y.

```python
#Using the matplotlib pyplot (plt)
x = np.sort(np.random.normal(size=100))
y = np.sort(np.random.normal(size=100))
plt.plot(x, y, 'ro')
```

```python
#Using pandas dataframes
mData = pd.DataFrame({'x':x,'y':y},columns=['x','y'])
mData.plot('x','y',kind='scatter')
```

## 2. Histogram
A histogram shows information about the distribution of a dataset.

```python
# Using matplotlib pyplot
vData = np.random.normal(size=10000, loc=40, scale=20)
plt.hist(vData,bins=100)
plt.show()
```

```python
# Using pandas dataframe
vDataFrame = pd.DataFrame(vData)
vDataFrame.hist(bins=100)
```

## 3. Boxplot
A boxplot also gives information about the distribution of the data. It shows the median and the spread of the data. More info on boxplots here: http://en.wikipedia.org/wiki/Box_plot

```python
# Using matplotlib pyplot
plt.boxplot(vData)
```

```python
# Using pandas dataFrame
vDataFrame.boxplot()
```
